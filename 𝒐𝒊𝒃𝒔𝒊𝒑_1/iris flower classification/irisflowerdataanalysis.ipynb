CLASSIFICATION

```python
#K nearest neighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

```

```python
iris_data
```

```python
dataframe=pd.DataFrame(iris_data)
```

```python
dataframe
```

```python
X=dataframe.drop('species',axis=1)
y=dataframe['species']
```

```python
X,y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
knnclf=KNeighborsClassifier(n_neighbors=3)
knnclf.fit(X_train,y_train)
y_pred=knnclf.predict(X_test)
```

EVALUATION


```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
cm=confusion_matrix(y_true=y_test,y_pred=y_pred)
cr=classification_report(y_true=y_test,y_pred=y_pred)
acs=accuracy_score(y_test,y_pred)
```

```python
print(cm)
```

```python
print(acs)
```

ANALYSIS
FOR VALUES OF K

```python
neighborlist=[]
a1list=[]
for i in range(1,50):
    knnclf=KNeighborsClassifier(n_neighbors=i)
    knnclf.fit(X_train,y_train)
    y_pred=knnclf.predict(X_test)
    a1=accuracy_score(y_test,y_pred)
    neighborlist.append(i)
    a1list.append(a1)
```

```python
print(neighborlist,a1list)
```

```python
plt.plot(a1list,neighborlist,marker='o')
plt.xlabel('Accuracy')
plt.ylabel('n_neighbors')
plt.title('Accuracy vs. n_neighbors for K-Nearest Neighbors')
plt.grid(True)
plt.show()

```
