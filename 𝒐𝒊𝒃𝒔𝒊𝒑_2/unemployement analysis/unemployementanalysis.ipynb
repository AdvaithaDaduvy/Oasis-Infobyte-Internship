```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (replace this with your actual DataFrame)
# Assuming df2 contains columns 'Date' and 'Estimated Unemployment Rate (%)'
# And df2['Estimated Unemployment Rate (%)'] contains the numerical data for the KDE


# Create kernel density plot
plt.figure(figsize=(8, 6))
plt.hist(df2['Estimated Unemployment Rate (%)'], bins=30, density=True, alpha=0.5, color='blue')  # Display histogram for reference
plt.title('Kernel Density Plot')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Density')
plt.grid(True)

# Kernel density estimate
kde = np.linspace(df2['Estimated Unemployment Rate (%)'].min(), df2['Estimated Unemployment Rate (%)'].max(), 1000)
kde_values = []
bandwidth = 0.5  # Adjust the bandwidth value as needed

for x in kde:
    kernel = (1 / (bandwidth * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((df2['Estimated Unemployment Rate (%)'] - x) / bandwidth)**2)
    kde_values.append(np.sum(kernel) / len(df2))

plt.plot(kde, kde_values, color='red', label='KDE')
plt.legend()
plt.show()
```
